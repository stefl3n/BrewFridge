```python
import pandas as pd
import matplotlib.pyplot as plt

# Path of the file to read
feeds_filepath = "./feeds.csv"

# Read the file into a variable data
data = pd.read_csv(feeds_filepath, index_col="created_at", parse_dates=True)

plt.figure(figsize=(16, 6))

# plotting the line 1 points
plt.plot(data['field1'], label = "Fridge")

# plotting the line 2 points
plt.plot(data['field5'], 'red', label = "External")

# naming the x axis
plt.xlabel('Time')
# naming the y axis
plt.ylabel('Temperature')
# giving a title to my graph
plt.title('Temperatures')

# show a legend on the plot
plt.legend()

# function to show the plot
plt.show()
```

```python
plt.figure(figsize=(16,2))

plt.plot(data['field3'], '#ffff00', label="Fridge Status (ON/OFF)")

plt.xlabel('Time')

plt.title('Fridge Status')

plt.yticks([0,1], ['OFF', 'ON'])

plt.show()
```
